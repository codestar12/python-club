```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# read 'csv' file into pandas dataframe

# update this to the location of the NHL 2015 file
df = pd.read_csv("/Users/codyblakeney/Documents/jupyterExample_v2/NHL 2015-16-main-page.csv")
```

```python
df # show data from
```

```python
countryList = []                        # create an empty list
countryList.append(sum(df.Nat=="USA"))  # add number of players from USA
countryList.append(sum(df.Nat=="CAN"))  # add number of players from Canada

# add players not from USA or Canada
countryList.append(sum(True for player in df.Nat if not(player == "USA" or
                                                        player == "CAN") ))
                                        
x = np.arange(len(countryList))
width = 1 / len(x)
```

```python
countryPlt = plt.bar(x, countryList)     # create a bar graph of players home countries

plt.xlabel('Country')                    # label x + y axes      
plt.ylabel('Number of Hockey Players')
plt.xticks(x + width, ('USA', 'Canada', 'Other'))

plt.show(countryPlt)                     # show the graph
```

```python
height = plt.boxplot(df.HT)
plt.ylabel("Height in Inches")
plt.show(height)
```

```python
salaryAge = plt.scatter(df.Age, df.Salary)
plt.ylabel("Salary")
plt.xlabel("Age")
plt.show(salaryAge)
```
